## Gerando uma animação

```python
# Coloque nesta lista os deslocamentos do robo

movimentos = [[-10, -10, 0], [-10, 10, 0], [-10,0,0], [-10, 0, 0],
              [0,0,math.pi/12.0], [0, 0, math.pi/12.0], [0, 0, math.pi/12],[0,0,-math.pi/4],
              [-5, 0, 0],[-5,0,0], [-5,0,0], [-10,0,0],[-10,0,0], [-10,0,0],[-10,0,0],[-10,0,0],[-15,0,0],
              [0,0,-math.pi/4],[0, 10, 0], [0,10,0], [0, 10, 0], [0,10,0], [0,0,math.pi/8], [0,10,0], [0,10,0], 
              [0,10,0], [0,10,0], [0,10,0],[0,10,0],
              [0,0,-math.radians(90)],
              [math.cos(math.pi/3)*10, math.sin(math.pi/3),0],[math.cos(math.pi/3)*10, math.sin(math.pi/3),0],[math.cos(math.pi/3)*10, math.sin(math.pi/3),0],
              [math.cos(math.pi/3)*10, math.sin(math.pi/3),0]
              
              ]
```

```python

plt.ioff() # Desliga o modo interativo, para nao aparecerem muitas imagens no meio

frames = 1

for delta in movimentos:
    for i in range(len(pose)):
        # Só para o robô andar
        # 0,1,2 são x,y,theta
        pose[i]+=delta[i]
        
    # particulas
    for j in b:


        # Desvio padrão aproximado em centímetros.

        std_x = 2
        std_y = 2  
        std_theta = 5



        erro_x = rd.gauss(0,std_x)
        erro_y = rd.gauss(0,std_y)
        erro_theta = rd.gauss(0,std_theta)

        desloc_x = delta[0]
        desloc_y = delta[1]
        desloc_theta = delta[2]

        j.x += desloc_x + erro_x
        j.y += desloc_y + erro_y
        j.theta += desloc_theta + erro_theta
        
        
            
    # Simula a leitura do lidar
    leituras, lidar_map = inspercles.nb_simulate_lidar(pose, angles, np_image)
    
    
    
    lista_produto =[]
    sigma =5

    

    for elementos in b:
        a = inspercles.nb_simulate_lidar([elementos.x,elementos.y,elementos.theta], angles, np_image)[0]
        produto=1
       
        for j in range (len (leituras)):    
            P = math.e**(-(leituras.keys()[0]-a.keys()[0])/2*(sigma**2))
            produto = produto * P
            
        lista_produto.append(produto)
                    
    soma = sum(lista_produto)
    alpha = 1/soma
    
    lista_final =[]
    
    #lp = np.array(lista_produto)
    #lp = alpha*lp
    
    for e in lista_produto:
        e = e*alpha
        lista_final.append(e)
        
    novas_particulas = Particle.draw_random_sample(b, lista_final, 500)
    
    novas_particulas_desv = []
    
    for part in novas_particulas:
        x_d = np.random.normal(-1, 1)
        y_d = np.random.normal(-1,1)
        theta_d = np.random.normal(-1,1)
        
        novas_particulas_desv.append([part.x+x_d, part.y+y_d, part.theta+theta_d])
    
    # Desenha as particulas
    ax = inspercles.nb_draw_map(color_image, pose=pose, robot=True, particles=novas_particulas_desv)
    #ax.imshow(occupancy_image, alph.2)
    # Desenha o mapa do lidar
    ax.imshow(lidar_map, alpha=0.5)

    plt.savefig("anim%04d.png"%frames, bounds="tight")
    
    frames+=1
    plt.close('all')


plt.ion()

```
